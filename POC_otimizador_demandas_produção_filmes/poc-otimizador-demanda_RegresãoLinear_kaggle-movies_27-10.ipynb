#Importando dependências

# Separando features/target e codificando target

```
X = dataset_fs.iloc[:,[0,1,2,3,4,6,7,8,9,10]]
X.head(3)
```

```
info_types=X.dtypes
info_types = info_types[info_types=='object']
info_types
```

```
le = LabelEncoder()

for i in info_types.index:
  X[i]= le.fit_transform(X.loc[:,i])

X.head(3)
```

```
X['budget']= st.zscore(X['budget'])
X
```

```
y = pd.DataFrame(st.zscore(dataset_fs['gross']), columns=['gross'])
y.head(3)
```

```
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=0, test_size=0.20)
```

# Indução do Modelo
 




### Usando SKlearn

```
lr = LinearRegression().fit(X_train, y_train)

print("Acurácia na base de treinamento: {:.2f}".format(lr.score(X_train, y_train)))
print("Acurácia na base de teste: {:.2f}".format(lr.score(X_test, y_test)))
```

```
dataset_proc =pd.concat([y,X],axis=1)
dataset_proc.head(2)
```

### Usando anova model

```
model = sm.formula.ols("gross ~	budget + company + country + director + genre + runtime + rating + score + star + writer", dataset_proc).fit()
sm.stats.anova_lm(model)
```
